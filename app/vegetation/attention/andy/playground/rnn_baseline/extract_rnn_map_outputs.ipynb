# Extract LFMC data from RNN maps

**Purpose:** We want to compare metrics of the prediction of the RNN maps to predictions by the transformer model. This notebook extracts LFMC from from the RNN maps.

**Date:** July 29, 2024

```python
import pandas as pd
from hydroDL import kPath
from datetime import datetime, timedelta
import rasterio
```

```python
sites_path = f'{kPath.dirVeg}model/data/singleDaily-nadgrid/info.csv'
sites_df = pd.read_csv(sites_path)
```

```python
# TODO: Code not intuitive
dates = []

start_date = datetime(2016, 4, 15)
end_date = datetime(2020, 10, 1)
current_date = start_date
while current_date <= end_date:
    if current_date.day == 15:
        dates.append(current_date.strftime('%Y-%m-%d'))
        current_date = current_date.replace(day=1) + timedelta(days=31)
        current_date = current_date.replace(day=1)
    else:
        dates.append(current_date.strftime('%Y-%m-%d'))
        current_date = current_date.replace(day=15)
```

```python
def get_lfmc(lat, lon, date):
   tif_path = f'{kPath.dirVeg}/maps/lfmc_map_{date}.tif'
   with rasterio.open(tif_path) as src:
      row, col = src.index(lon, lat)
      lfmc_val = src.read(1)[row, col]
   return lfmc_val
```

```python
for i, site in enumerate(sites_df.iterrows()):
    print(f'Site {i} / {len(sites_df)}')
    site_id = site[1].siteId
    lat = site[1].lat
    lon = site[1].lon

    site_dates = []
    site_lfmc_vals = []
    for date in dates:
        try:
            lfmc_val = get_lfmc(lat, lon, date)
            
            site_dates.append(date)
            site_lfmc_vals.append(lfmc_val)   
        
        except:
            print(date, 'does not exist.')
        
    site_data = {'lfmc' : site_lfmc_vals}
    df = pd.DataFrame(site_data)
    df = df.set_index(pd.DatetimeIndex(site_dates))

    out_path =f'{kPath.dirVeg}inference/rnn_maps/{site_id}.csv'
    df.to_csv(out_path)
```
