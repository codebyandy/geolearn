# Load data

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

folder = r'C:\Users\geofk\work\waterQuality\tempData\Lijing'
data = np.load(os.path.join(folder, 'data.npz'))
dfG = pd.read_csv(os.path.join(folder, 'geoConst'))
```


plot precipitation of the 10th site

```python
%matplotlib inline
t = data['t']
varF = list(data['varF'])
p = data['f'][:, 10, varF.index('pr')]
fig, ax = plt.subplots(1, 1)
ax.plot(t, p, '-b')
fig.show()
```

plot Ca concentration for the 10th site 

```python
%matplotlib inline
t = data['t']
varC = list(data['varC'])
p = data['c'][:, 10, varC.index('00915')]
fig, ax = plt.subplots(1, 1)
ax.plot(t, p, '*')
fig.show()
```
